```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.relplot(x = "total_bill", y = "tip", hue= "smoker", data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", ci = None, data = df)
```

```python
sns.lmplot(x = "total_bill", y = "tip", 
           hue = "smoker", col = "sex", ci = None, data = df);
```

```python
# Anzahl der Raucher und nicht Raucher 
smoker_counts=df["smoker"].value_counts()
smoker_counts
```

```python
plt.figure(figsize=(5, 5))
plt.pie(smoker_counts, 
        labels=smoker_counts.index, 
        autopct='%1.1f%%', 
        colors=['lightcoral', 'lightskyblue'], 
        startangle=90)
plt.title('Verteilung der Raucher im Datensatz Tips')
plt.show()
```

```python
day_counts = df['day'].value_counts()
day_counts
```

```python
plt.figure(figsize=(5, 5))
plt.pie(day_counts, 
        labels=day_counts.index, 
        autopct='%1.1f%%', 
        colors=['lightcoral', 'lightskyblue', 'lightyellow', 'lightgreen'], 
        startangle=90)
plt.title('Wochentage im Datensatz Tips')
plt.show()
```
